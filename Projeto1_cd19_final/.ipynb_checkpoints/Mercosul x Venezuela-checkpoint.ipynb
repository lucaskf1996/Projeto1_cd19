## Com as políticas de livre comércio entre os países do MERCOSUL, a Venezuela fez alguma diferença nos outros países do bloco econômico?

Espera-se que haja um incremento nas quantidades de importação e exportação dos outros países integrantes do bloco econômico devido as políticas de livre comércio que isentam taxas dos produtos.

Para esta análise, serão avaliados as variações do PIB, exportações (% do PIB) e importações (% do PIB) de todos os 4 países do Mercado Comum do Sul e da Venezuela, com um foco principal nos anos próximos de 2006 e 2012, anos de entrada e suspensão da Venezuela no bloco, além de algumas análises gerais utilizando períodos de tempo mais distantes e outros países de outros blocos para adicionar um contexto econômico mais global.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd

export = pd.read_excel("export.xls")
```

```python
export_indexado = export.set_index("exports")
export_brazil = export_indexado.loc["Brazil"]
years = []
for i in export.columns:
    if i != "exports":
        years.append(int(i)) #tentativa falha de transformar os anos em int
export_brazil
```

```python
export_brazil.plot(grid = True, figsize=(20,10))
```

## Exportação
### MERCOSUL

```python
export_mercosul = export_indexado.loc[["Brazil", "Argentina", "Paraguay", "Uruguay"]]
export_mercosul = export_mercosul.transpose()
export_mercosul.index = export_mercosul.index.astype(dtype = "int64")
export_mercosul.plot(grid = True, figsize=(20,10))
```

### MERCOSUL no período de integração e suspensão da Venezuela

```python
export_2006_2012 = export_mercosul.loc[[2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]]
export_2006_2012.plot(grid = True, figsize=(20,10))
```

### Somente Brasil e Argentina durante a criação do MERCOSUL

```python
export_bra_arg_1991 = export_mercosul.loc[[1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995],["Brazil","Argentina"]]
export_bra_arg_1991.plot(grid = True, figsize=(20,10))
```

### Somente Brasil e Argentina durante a criação do MERCOSUL

```python
export_par_uru_1991 = export_mercosul.loc[[1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995],["Paraguay", "Uruguay"]]
export_par_uru_1991.plot(grid = True, figsize=(20,10))
```

## PIB
### Tabela

```python
gdp = pd.read_excel("gdp.xls")
gdp_indexado = gdp.set_index("Country Name")
gdp_mercosul = gdp_indexado.loc[["Brazil", "Argentina", "Paraguay", "Uruguay"]]
gdp_mercosul
```

### MERCOSUL

```python
gdp_mercosul = gdp_mercosul.transpose()
gdp_mercosul.plot(grid = True, figsize=(20,10))
```

### MERCOSUL no período de integração e suspensão da Venezuela

```python
gdp_2006_2012 = gdp_mercosul.loc[[2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]]
gdp_2006_2012.plot(grid = True, figsize=(20,10))
```

### Somente Paraguai e Uruguai (para melhor visualização)

```python
gdp_paraguay_uruguay = gdp_indexado.loc[["Uruguay", "Paraguay"]]
gdp_paraguay_uruguay = gdp_paraguay_uruguay.transpose()
gdp_paraguay_uruguay.plot(grid = True, figsize=(20,10))
```

### Paraguai e Uruguai no período de integração e suspensão da Venezuela

```python
gdp_par_uru_2006_2012 = gdp_paraguay_uruguay.loc[[2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]]
gdp_par_uru_2006_2012.plot(grid = True, figsize=(20,10))
```

## Importação

### MERCOSUL

```python
importacao = pd.read_excel("importacao.xls")
importacao_indexado = importacao.set_index("Country Name")
importacao_mercosul = importacao_indexado.loc[["Brazil", "Paraguay", "Uruguay", "Argentina"]]
importacao_mercosul = importacao_mercosul.transpose()
importacao_mercosul.index = importacao_mercosul.index.astype(dtype = "int64")
importacao_mercosul.plot(grid = True, figsize=(20,10))
```

### MERCOSUL durante a integração e suspensão da Venezuela

```python
importacao_mercosul_2006_2012 = importacao_mercosul.loc[[2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]]
importacao_mercosul_2006_2012.plot(grid = True, figsize=(20,10))
print("O gráfico aparenta seguir a estabilização antes da entrada")
```

### Somente Paraguai e Uruguai

```python
importacao_par_uru_1991 = importacao_mercosul.loc[:, ["Paraguay", "Uruguay"]]
importacao_par_uru_1991.plot(grid = True, figsize=(20,10))
```

### Somente Brasil e Argentina

```python
importacao_bra_arg_1991 = importacao_mercosul.loc[:, ["Brazil", "Argentina"]]
importacao_bra_arg_1991.plot(grid = True, figsize=(20,10))
```

As importações do Brasil provavelmente foram causadas por conta das políticas do presidente Fernando Collor
